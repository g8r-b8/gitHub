```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 

data = pd.read_csv('credit_data.csv') 
```

```python
print(type(data))
data
```

```python
replacements = {
    'good':1,
    'bad':2
}
data['class'].replace(replacements, inplace=True)
data
```

```python
new_data = data[['housing','current_balance','location','cc_age','existing_credits','num_dependents','class']]
replacements_x = {
    'own':1,
    'for free':2,
    'rent':3
}
new_data['housing'].replace(replacements_x, inplace=True)
new_data
```

```python

x = new_data.iloc[:, range(0,6)].values  
print(x) 
y = new_data.iloc[:, 6].values   
for e in y:
    print(e)
```

```python
from sklearn.linear_model import LogisticRegression
bnb = LogisticRegression()
y=y.astype('int')
```

```python
bnb.fit(x,y)
```

```python
new_data.info()
new_data.to_csv()
```

```python
bnb.score(x,y)
```

```python
x1test=[[1 ,2050 ,2,49,1,1]]

ynew = bnb.predict(x1test)
for e in ynew:
    print(e)
```

```python
x2test=x
y2new=bnb.predict(x2test)
for e in y2new:
    print(e)
```
